```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
alturas = np.array([1.45, 1.7, 1.60, 1.8, 1.95, 1.73, 1.50, 1.88, 1.89, 1.77, 1.55, 1.91]).reshape(-1, 1)
pesos = np.array([64,71, 99, 81, 91, 74, 67, 88, 103, 73, 55,100]).reshape(-1, 1)

alturas_teste = np.array([1.71, 1.80, 1.61]).reshape(-1,1)
pesos_teste = np.array([74, 87, 70]).reshape(-1,1)
```

```python
plt.scatter(alturas, pesos)
plt.grid(True)
plt.title("Altura x Peso")
plt.ylabel("Peso (Kg)")
plt.xlabel("Altura (Metros)")
plt.show()
```

```python
#Importação do modelo KNN e Linear
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
```

```python
# Criando os modelos
knn_1 = KNeighborsRegressor(n_neighbors = 1).fit(alturas, pesos)
knn_3 = KNeighborsRegressor(n_neighbors = 3).fit(alturas, pesos)
knn_5 = KNeighborsRegressor(n_neighbors = 5).fit(alturas, pesos)
linear = LinearRegression().fit(alturas, pesos)
```

```python
# Exibindo os gráficos
# Exibindo as linhas de predição
#legendas = ["Linha de prediçao", "Treino", "Teste"]
figura, eixos = plt.subplots(2, 2, figsize=(10,10))


# Código para criar a linha de predição dos dados de treino
# onde é criado 500 valores igualmente espaçados entre 1.45 e 1.95
linha = np.linspace(1.45, 1.95, 1000).reshape(-1, 1)

#Knn 1
eixo = eixos[0,0]
eixo.grid(True)
eixo.plot(linha, knn_1.predict(linha))
eixo.plot(alturas, pesos, "^", markersize=5)
eixo.plot(alturas_teste, pesos_teste, "v", markersize=8)
eixo.set_title("KNN {}\n Treino: {:.2f} - Teste: {:.2f}".format(1, knn_1.score(alturas, pesos), knn_1.score(alturas_teste, pesos_teste)))

#Knn 3
eixo = eixos[0,1]
eixo.grid(True)
eixo.plot(linha, knn_3.predict(linha))
eixo.plot(alturas, pesos, "^", markersize=5)
eixo.plot(alturas_teste, pesos_teste, "v", markersize=8)
eixo.set_title("KNN {}\n Treino: {:.2f} - Teste: {:.2f}".format(3, knn_3.score(alturas, pesos), knn_3.score(alturas_teste, pesos_teste)))

#Knn 5
eixo = eixos[1,0]
eixo.grid(True)
eixo.plot(linha, knn_5.predict(linha))
eixo.plot(alturas, pesos, "^", markersize=5)
eixo.plot(alturas_teste, pesos_teste, "v", markersize=8)
eixo.set_title("KNN {}\n Treino: {:.2f} - Teste: {:.2f}".format(5, knn_5.score(alturas, pesos), knn_5.score(alturas_teste, pesos_teste)))

#Linear
eixo = eixos[1,1]
eixo.grid(True)
eixo.plot(linha, linear.predict(linha))
eixo.plot(alturas, pesos, "^", markersize=5)
eixo.plot(alturas_teste, pesos_teste, "v", markersize=8)
eixo.set_title("Modelo Linear\n Treino: {:.2f} - Teste: {:.2f}".format(linear.score(alturas, pesos), linear.score(alturas_teste, pesos_teste)))

#eixos[0].legend(legendas)

plt.show()
```

```python
# Valores de Predição
valores = pesos_teste
print("Pesos Teste: {}, {}, {}".format(valores[0][0], valores[1][0], valores[2][0]))
valores = knn_1.predict(alturas_teste)
print("Knn {} - {}, {}, {}".format(1, valores[0][0], valores[1][0], valores[2][0]))
valores = knn_3.predict(alturas_teste)
print("Knn {} - {}, {}, {}".format(3, valores[0][0], valores[1][0], valores[2][0]))
valores = knn_5.predict(alturas_teste)
print("Knn {} - {}, {}, {}".format(5, valores[0][0], valores[1][0], valores[2][0]))
valores = linear.predict(alturas_teste)
print("Modelo Linear - {}, {}, {}".format(valores[0][0], valores[1][0], valores[2][0]))
```
